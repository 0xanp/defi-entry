```python
################################
# Libraries
################################
import pandas as pd
import seaborn as sns; sns.set_theme()
```

```python
# Loading and concat dataframes
csv1 = pd.read_csv('./uniswap_13513629_13613629.csv',usecols = ['balance','value','gas','gasUsed','gasPrice'])
csv2 = pd.read_csv('./uniswap_13613629_13713629.csv',usecols = ['balance','value','gas','gasUsed','gasPrice'])
data = pd.concat([csv1, csv2], ignore_index=True) 
# Interpret columns as numerics (int64 or float64) for column operations
data = data.apply(pd.to_numeric,errors='coerce')
# Adding txn_percent column
data['txn_percent'] = (data['gasPrice']/data['value'])*100
```

```python
#Set Constants
# 1000 USD ~ .24 ether
SMALL_ACCOUNT_LIMIT = 24 * 1e16
# 1 000 000 USD ~ 240 ether
TIER_2_LIMIT = 240 * 1e18
#Splitting data into tiers
tier1 = data[data['balance'] < SMALL_ACCOUNT_LIMIT]
tier2 = data[data['balance'] >= SMALL_ACCOUNT_LIMIT]
tier2 = tier2[tier2['balance'] < TIER_2_LIMIT]
tier3 = data[data['balance'] >= TIER_2_LIMIT]
```

```python
#Find correlation between tier1 metrics
tier1_corr = tier1[['balance','value','gas','gasUsed','gasPrice','txn_percent']].corr(method='pearson')
tier1_corr
```

```python
# Correlation heatmap for tier1 accounts
ax = sns.heatmap(tier1_corr)
```
