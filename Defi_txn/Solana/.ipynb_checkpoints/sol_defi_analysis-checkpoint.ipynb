```python
################################
# Libraries
################################
import pandas as pd
import seaborn as sns; sns.set_theme()
```

```python
# Load data, interpret columns into numeric types (int64 or float64) for column operations and extract non-0 transactions
csv1 = pd.read_csv('./defi_txn1.csv', encoding= 'unicode_escape')
csv2 = pd.read_csv('./defi_txn2.csv', encoding= 'unicode_escape')
data = pd.concat([csv1, csv2], ignore_index=True)
data.drop_duplicates(subset=['sig'])
data['transacted'] = pd.to_numeric(data['transacted'],errors='coerce')
data['fee'] = pd.to_numeric(data['fee'],errors='coerce')
data['balance'] = pd.to_numeric(data['balance'],errors='coerce')
data = data[data['transacted'] > 0]
# Adding txn_percent column
data['fee_percent'] = (data['fee']/data['transacted'])*100
```

```python
#Splitting data into tiers
##########################################
# Data description
# - Tier1 accounts - 1 <= bal <= 1000 USD 
# - Tier2 accounts - 1000 < bal <= 1000000 USD
# - Tier3 accounts - 1000000 USD < bal
# - Balances are in lamport = 1^10e-9 SOL 
##########################################
tier1 = data[data['balance']<=4670000000]
tier2 = data[data['balance']>4670000000]
tier2 = tier2[tier2['balance']<=4670000000000]
tier3 = data[data['balance']>4670000000000]
```

```python
tier1.describe()
```

```python
###################################
# Correlation between tier1 metrics
###################################
# Correlation coefficient table for small accounts
tier1_corr = tier1[['balance','transacted','fee','fee_percent']].corr(method='pearson')
tier1_corr
```

```python
# Correlation heatmap for tier1 accounts
ax = sns.heatmap(tier1_corr)
```

```python
###################################
# Correlation between tier2 metrics
###################################
# Correlation coefficient table for small accounts
tier2_corr = tier2[['balance','transacted','fee','fee_percent']].corr(method='pearson')
tier2_corr
```

```python
# Correlation heatmap for tier2 accounts
ax = sns.heatmap(tier2_corr)
```

```python
###################################
# Correlation between tier3 metrics
###################################
# Correlation coefficient table for small accounts
tier3_corr = tier3[['balance','transacted','fee','fee_percent']].corr(method='pearson')
tier3_corr
```

```python
# Correlation heatmap for tier3 accounts
ax = sns.heatmap(tier3_corr)
```
