```python
################################
# Libraries
################################
import pandas as pd
import seaborn as sns; sns.set_theme()
```

```python
#Set Constants
SMALL_ACCOUNT_LIMIT = 2 * 1e18
TIER_2_LIMIT = 2000 * 1e18
```

```python
#Load Data
router_data = pd.read_csv('router_data_with_balances')
router_data = router_data[router_data['Value_IN(BNB)']>0]
```

```python
#add fee as % of txn and $ of account 
router_data['fee_as_prcnt_of_tx'] =   router_data['TxnFee(BNB)'] / router_data['Value_IN(BNB)']
router_data['fee_as_prcnt_of_account'] = router_data['TxnFee(BNB)'] / router_data['from_balance']
router_data
```

```python
tier1 = router_data[router_data['from_balance'] < SMALL_ACCOUNT_LIMIT]
tier2 = router_data[router_data['from_balance'] >= SMALL_ACCOUNT_LIMIT]
tier2 = tier2[tier2['from_balance'] < TIER_2_LIMIT]
tier3 = router_data[router_data['from_balance'] >= TIER_2_LIMIT]
```

```python
tier1.describe()
```

```python
tier2.describe()
```

```python
tier3.describe()
```

```python
#Find correlation between tier1 metrics
tier1_corr = tier1[['Value_IN(BNB)', 'TxnFee(BNB)', 'TxnFee(USD)', 'from_balance', 'fee_as_prcnt_of_tx', 'fee_as_prcnt_of_account']].corr(method='pearson')
tier1_corr
```

```python
# Correlation heatmap for tier1 accounts
ax = sns.heatmap(tier1_corr)
```

```python
#Find correlation between tier2 metrics
tier2_corr = tier2[['Value_IN(BNB)', 'TxnFee(BNB)', 'TxnFee(USD)', 'from_balance', 'fee_as_prcnt_of_tx', 'fee_as_prcnt_of_account']].corr(method='pearson')
tier2_corr
```

```python
# Correlation heatmap for tier2 accounts
ax = sns.heatmap(tier2_corr)
```

```python
#Find correlation between tier2 metrics
tier3_corr = tier3[['Value_IN(BNB)', 'TxnFee(BNB)', 'TxnFee(USD)', 'from_balance', 'fee_as_prcnt_of_tx', 'fee_as_prcnt_of_account']].corr(method='pearson')
tier3_corr
```

```python
# Correlation heatmap for tier3 accounts
ax = sns.heatmap(tier3_corr)
```
