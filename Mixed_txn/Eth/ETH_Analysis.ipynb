```python
################################
# Libraries
################################
import pandas as pd
import seaborn as sns; sns.set_theme()
```

```python
#Import ETH Data
large = pd.read_csv('large.csv')
large['balance'] = large['balance'].astype('float64')
medium = pd.read_csv('medium.csv')
medium['balance'] = medium['balance'].astype('float64')
small = pd.read_csv('small.csv')
small['balance'] = small['balance'].astype('float64')
```

```python
#Build Correlations
large_corr = large[['balance', 'num_txn', 'avg_tx_value', 'avg_gas', 'avg_gas_used', 'avg_gas_price', 'avg_tx_fee', 'avg_tx_pct']].corr(method='pearson')
large_corr
```

```python
#Build Correlations
medium_corr = medium[['balance', 'num_txn', 'avg_tx_value', 'avg_gas', 'avg_gas_used', 'avg_gas_price', 'avg_tx_fee', 'avg_tx_pct']].corr(method='pearson')
medium_corr
```

```python
#Build Correlations
small_corr = small[['balance', 'num_txn', 'avg_tx_value', 'avg_gas', 'avg_gas_used', 'avg_gas_price', 'avg_tx_fee', 'avg_tx_pct']].corr(method='pearson')
small_corr
```

```python
ax = sns.heatmap(large_corr)
```

```python
ax = sns.heatmap(medium_corr)
```

```python
ax = sns.heatmap(small_corr)
```
