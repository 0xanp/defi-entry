```python
################################
# Libraries
################################
import pandas as pd
import seaborn as sns; sns.set_theme()
```

```python
################################
# Loading data
################################
tier1 = pd.read_csv('./sol_tier1_accounts.csv')
tier2 = pd.read_csv('./sol_tier2_accounts.csv')
tier3 = pd.read_csv('./sol_tier3_accounts.csv')
```

```python
##########################################
# Data description
# - Tier1 accounts - 1 <= bal <= 1000 USD 
# - Tier2 accounts - 1000 < bal <= 1000000 USD
# - Tier3 accounts - 1000000 USD < bal
# - Balances are in lamport = 1^10e-9 SOL 
##########################################
tier1.describe()
```

```python
tier2.describe()
```

```python
tier3.describe()
```

```python
############################################
# Some special statistics in all three tiers
# Heavy traders - more than 100 transactions
# HODLers - less than 3 transactions
############################################
# Tier1
print("Number of heavy traders in Tier 1 (more than 100 transactions): ", len(tier1[tier1['txn_count']>=100]))
print("Number of HODLers in Tier 1 (less than 5 transactions): ", len(tier1[tier1['txn_count']<=3]))
# Tier2
print("Number of heavy traders in Tier 2 (more than 100 transactions): ", len(tier2[tier2['txn_count']>=100]))
print("Number of HODLers in Tier 2 (less than 5 transactions): ", len(tier2[tier2['txn_count']<=3]))
# Tier3
print("Number of heavy traders in Tier 3 (more than 100 transactions): ", len(tier3[tier3['txn_count']>=100]))
print("Number of HODLers in Tier 3 (less than 5 transactions): ", len(tier3[tier3['txn_count']<=3]))
```

```python
###################################
# Correlation between tier1 metrics
###################################
# Correlation coefficient table for small accounts
tier1_corr = tier1[['balance','txn_count','total_transacted','total_fee','avg_txn','avg_fee_percent']].corr(method='pearson')
tier1_corr
```

```python
# Correlation heatmap for tier1 accounts
ax = sns.heatmap(tier1_corr)
```

```python
###################################
# Correlation between tier2 metrics
###################################
# Correlation coefficient table for small accounts
tier2_corr = tier2[['balance','txn_count','total_transacted','total_fee','avg_txn','avg_fee_percent']].corr(method='pearson')
tier2_corr
```

```python
# Correlation heatmap for tier2 accounts
ax = sns.heatmap(tier2_corr)
```

```python
###################################
# Correlation between tier3 metrics
###################################
# Correlation coefficient table for small accounts
tier3_corr = tier3[['balance','txn_count','total_transacted','total_fee','avg_txn','avg_fee_percent']].corr(method='pearson')
tier3_corr
```

```python
# Correlation heatmap for tier3 accounts
ax = sns.heatmap(tier3_corr)
```

```python
#################################################################################
# Remarks about the correlation tables
# - In all tiers, there's no correlation between avg_fee_percent and key metrics
# like total_transacted, balance, and txn_count.
# - We can conclude that gas fee on Solana does not affect behaviors of users at 
# any tiers. Hence, the low balance accounts are not priced out.
#################################################################################
```
